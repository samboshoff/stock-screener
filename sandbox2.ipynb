```python
def filter_on_criteria(df: pd.DataFrame) -> pd.DataFrame:

    filtered_df = df[(df['pc_sma10_by_close']<0) & df['pc_sma30_by_sma100']>0]
    filtered_df = filtered_df.sort_values('pc_sma10_by_close')
    
    return filtered_df
```
